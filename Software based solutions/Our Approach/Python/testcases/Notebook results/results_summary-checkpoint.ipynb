## Test case 1

Number of unknowns : 2

Domain : Circuits

```python
"""
Easy circuit [diode and resistor] equations

Num unknown : 2
"""

import sys

sys.path.insert(0, '../../nn_python')
sys.path.insert(0, '../../ckt_simulator')
sys.path.insert(0, '../../parser')

import FFNN
import lib.ahkab as ahkab
from parser_class import Parser
import activation_functions as afs
from nn_trainer_class import Trainer
import numpy as np
from scipy.sparse import csc_matrix

### Script arguments
NUM_UNKNOWNS = 2
NUM_EXP = 1
NETLIST_NAME = "ckt_netlist_1.txt"


### Run the simulation and export the output files
CKT = ahkab.netlist_parser.parse_circuit(NETLIST_NAME)[0]
OPA = ahkab.new_op()
R = ahkab.run(CKT, OPA)['op']

ckt_res = []
for i in R.keys():
   ckt_res.append(R[i][0])
np.savetxt('ckt_res_1.txt', ckt_res)
#########

## Parser section
# create a parser to parse an eqn with NUM_UNKNOWNS unknwons
P = Parser(NUM_UNKNOWNS)
P.add_nonlin_neuron('exp', NUM_EXP)
P.parse()
########

## Neural network section
IN_VECT = np.array([[1]])
NET = FFNN.FFNN([1, NUM_UNKNOWNS, NUM_UNKNOWNS+NUM_EXP, NUM_UNKNOWNS], manh=True, eta=0.1)
OUT_VECT = np.zeros([NUM_UNKNOWNS, 1])

# first hidden layer initillization
NET.layers[0].weight_matrix = np.zeros([NUM_UNKNOWNS, 2])
NET.layers[0].with_bias = False # no bias for this layer
NET.layers[0].f_act = afs.identity
NET.layers[0].f_act_d = afs.identity_d

# second hidden layer initiallization
# Now set the 2nd hidden layer such that its weights are not changed during iterations
NET.layers[1].trainable = False
# Adjust the weights of this layer
NET.layers[1].weight_matrix = np.loadtxt('out_l_0.txt')
# each neuron has its own f_act
NET.layers[1].act_per_neuron = True
NET.layers[1].custom_f_act = [afs.identity]*NUM_UNKNOWNS + [afs.diode_exp]*NUM_EXP
NET.layers[1].custom_f_act_d = [afs.identity_d]*NUM_UNKNOWNS + [afs.diode_exp]*NUM_EXP
NET.layers[1].with_bias = False

# third hidden layer initiallization
NET.layers[2].trainable = False
# Adjust the weights of this layer
NET.layers[2].weight_matrix = np.loadtxt('out_l_1.txt')

# All the output neurons are with identity f_act
NET.layers[2].f_act = afs.identity
NET.layers[2].f_act_d = afs.identity_d
########

# Create solver object
solver = Trainer(NET, OUT_VECT)
# Invoke the trainer [manhatten]
solver.train_manh(IN_VECT, max_iter=18000, tol_max=2, stop_after=5000)


# Invoke the trainer [Adam]
solver.train_adam(IN_VECT, max_iter=18000, eta=0.001)


solver.plot_err()
```

## Test case 8 continued
Number of unknowns : 2000

Domain : Badly scaled Logarithmic function

> A Benchmarking Problem

One of our contributions to the neural network architecture is the scale layer, such that a dummy layer with a single trainable weight is introduced in order to compensate for the large deviation between the roots. 

**Notice how the above method failed to solve this problem and the error is very high** `1e3`

```python
"""
Logarithmic function testcases
Num unknowns : 2000
"""

import sys

sys.path.insert(0, '../../nn_python')
sys.path.insert(0, '../../ckt_simulator')
sys.path.insert(0, '../../parser')

import pylab
import time
import FFNN
import lib.ahkab as ahkab
from parser_class import Parser
import activation_functions as afs
from nn_trainer_class import Trainer
import numpy as np
from scipy.sparse import csc_matrix

### Script arguments
NUM_UNKNOWNS = 2001
NUM_EXP = 2000


## Parser section
# create a parser to parse an eqn with NUM_UNKNOWNS unknwons
P = Parser(NUM_UNKNOWNS, in_file='eqns8.txt')
P.add_nonlin_neuron('lg', NUM_EXP)
P.parse()
########

## Neural network section
IN_VECT = np.array([[1]])
NET = FFNN.FFNN([1, 1, NUM_UNKNOWNS, NUM_UNKNOWNS+NUM_EXP, NUM_UNKNOWNS], manh=True, eta=01.1)
OUT_VECT = np.zeros([NUM_UNKNOWNS, 1])

NET.layers[0].weight_matrix = 1*np.ones([1, 2])
NET.layers[0].with_bias = False
NET.layers[0].f_act = afs.identity
NET.layers[0].f_act_d = afs.identity_d
#NET.layers[0].trainable = False

# first hidden layer initillization
NET.layers[1].weight_matrix = 1*np.ones([NUM_UNKNOWNS, 2])
NET.layers[1].weight_matrix[-1,-1] = 4
NET.layers[1].with_bias = False # no bias for this layer
NET.layers[1].f_act = afs.identity
NET.layers[1].f_act_d = afs.identity_d

# second hidden layer initiallization
# Now set the 2nd hidden layer such that its weights are not changed during iterations
NET.layers[2].trainable = False
# Adjust the weights of this layer
NET.layers[2].weight_matrix = csc_matrix(np.loadtxt('out_l_0.txt'))
# each neuron has its own f_act
NET.layers[2].act_per_neuron = True
NET.layers[2].custom_f_act = [afs.identity]*NUM_UNKNOWNS + [afs.log_sim]*NUM_EXP
NET.layers[2].custom_f_act_d = [afs.identity_d]*NUM_UNKNOWNS + [afs.log_d_sim]*NUM_EXP
NET.layers[2].with_bias = False
NET.layers[2].parallel_fact = NUM_UNKNOWNS

# third hidden layer initiallization
NET.layers[3].trainable = False
# Adjust the weights of this layer
NET.layers[3].weight_matrix = csc_matrix(np.loadtxt('out_l_1.txt'))

# All the output neurons are with identity f_act
NET.layers[3].f_act = afs.identity
NET.layers[3].f_act_d = afs.identity_d
########

# Create solver object
solver = Trainer(NET, OUT_VECT)

for i in range(70):
    # Invoke the trainer [manhatten]
    solver.train_manh(IN_VECT, max_iter=800, tol_max=2, stop_after=5000)

    # Invoke the trainer [Adam]
    solver.train_adam(IN_VECT, max_iter=1800, eta=0.001)
solver.plot_err()
```

After introducing the scale layer, we test the network without this layer and compare the results when the scale layer is introduced. Below the scale layer is de-activated and the neural network is the same as the one trained in the previous example.

```python
solver = Trainer(NET, OUT_VECT)
NET.layers[0].trainable = False
NET.layers[0].weight_matrix = 1*np.ones([1, 2])

NET.layers[1].weight_matrix = 1*np.ones([NUM_UNKNOWNS, 2])
for i in range(70):
    # Invoke the trainer [manhatten]
    solver.train_manh(IN_VECT, max_iter=800, tol_max=2, stop_after=5000)

    # Invoke the trainer [Adam]
    solver.train_adam(IN_VECT, max_iter=1800, eta=0.001)
solver.plot_err()
```

Noice the error is almost `1e2` which is considered very bad

Now we introduce the **scale layer** into the neural network [**One of our contributions** in the neural network architecture]

```python
solver = Trainer(NET, OUT_VECT)
NET.layers[0].trainable = True
NET.layers[0].weight_matrix = 1*np.ones([1, 2])

NET.layers[1].weight_matrix = 1*np.ones([NUM_UNKNOWNS, 2])
for i in range(180):
    # Invoke the trainer [manhatten]
    solver.train_manh(IN_VECT, max_iter=500, tol_max=2, stop_after=5000)

    # Invoke the trainer [Adam]
    solver.train_adam(IN_VECT, max_iter=900, eta=0.001)
solver.plot_err()
```
